```python
#Importing all libraries
import numpy as np

import matplotlib.pyplot as plt

import pandas as pd
```

```python
#Importing DataSet
data=pd.read_csv('/content/drive/MyDrive/PROJECT/databook csv.csv')

x=data.iloc[:,:1].values  #Features

y=data.iloc[:,5:8].values #Dependent variables
```

```python
print("--------TEMPARATURE VALUES AS FEATURE------------\n")

print("[   T  ]")

print(x)
```

```python
print("--------Prandtl number , Specific heat and Thermal conductivity as Dependent variables------------\n")

print("[    Pr         Cp        K")

print(y)
```

```python
#Splitting the dataset into Train data and Test data
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
print("---------X TRAIN----------\n")

print(x_train)
```

```python
print("---------X Test----------\n")

print(x_test)
```

```python
print("----------Y TRAIN----------\n")

print(y_train)
```

```python
print("---------Y TEST----------\n")

print(y_test)
```

```python
#Linear Regressiom
from sklearn.linear_model import LinearRegression

regressor=LinearRegression()

#Polynomial Regression
from sklearn.preprocessing import PolynomialFeatures

poly_reg=PolynomialFeatures(degree=4)

x_poly=poly_reg.fit_transform(x_train)

regressor.fit(x_poly,y_train)
```

```python
y_pred=regressor.predict(poly_reg.transform(x_test))

print("------PRANDTL NUMBER--------\n")

np.set_printoptions(precision=3)

print("[ TEST  PREDICTED]")

print()

print(np.concatenate((y_test[:,0].reshape(len(y_test[:,0]),1),y_pred[:,0].reshape(len(y_pred[:,0]),1)),1))

print("\n-------r2 score---------\n")

from sklearn.metrics import r2_score

r2_score(y_test[:,0],y_pred[:,0])
```

```python
yy = regressor.predict(poly_reg.fit_transform(x_train))

plt.title("Temparature vs Prandtl Number")

plt.scatter(x_train,y_train[:,0],color = "blue")

plt.scatter(x_train,yy[:,0],color = 'red')

plt.xlabel("Temparature")

plt.ylabel("Prandtl Number")

plt.show()
```

```python
print("------SPECIFIC HEAT--------\n")

np.set_printoptions(precision=3)

print("[ TEST    PREDICTED]")

print()

print(np.concatenate((y_test[:,1].reshape(len(y_test[:,1]),1),y_pred[:,1].reshape(len(y_pred[:,1]),1)),1))

print("\n-------r2 score---------\n")

r2_score(y_test[:,1],y_pred[:,1])
```

```python
yy = regressor.predict(poly_reg.fit_transform(x_train))

plt.title("Temparature vs Specific heat ")

plt.scatter(x_train,y_train[:,1],color = "blue")

plt.scatter(x_train,yy[:,1],color = 'red')

plt.xlabel("Temparature")

plt.ylabel("Specific Heat")

plt.show()
```

```python
print("------THERMAL CONDUCTIVITY--------\n")

np.set_printoptions(precision=3)

print("[ TEST    PREDICTED]")

print()

print(np.concatenate((y_test[:,1].reshape(len(y_test[:,1]),1),y_pred[:,1].reshape(len(y_pred[:,1]),1)),1))

print("\n-------r2 score---------\n")

r2_score(y_test[:,2],y_pred[:,2])
```

```python
yy = regressor.predict(poly_reg.fit_transform(x_train))

plt.title("Temparature vs Thermal Conductivity")

plt.scatter(x_train,y_train[:,2],color = "blue")

plt.scatter(x_train,yy[:,2],color = 'red')

plt.xlabel("Temparature")

plt.ylabel("Thermal Conductivity")

plt.show()
```
